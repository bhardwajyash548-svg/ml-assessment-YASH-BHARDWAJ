# 1. Data Loading & Inspection

```python


import pandas as pd

# Load dataset
df = pd.read_csv('../data/q1_heart_disease.csv')

# Basic inspection
print("Shape:", df.shape)
print("\nData Types:\n", df.dtypes)
print("\nMissing Values:\n", df.isnull().sum())

# First 5 rows
df.head()
```

# TASK-2 Exploratory Data Analysis

# (a) Target Distribution

```python


import matplotlib.pyplot as plt
import seaborn as sns

sns.countplot(x='heart_disease', data=df)
plt.title("Target Variable Distribution")
plt.show()
```

# (b) Correlation Heatmap

```python


df_encoded = pd.get_dummies(df, drop_first=True)

plt.figure(figsize=(12,8))
sns.heatmap(df_encoded.corr(), cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()

# Correlation heatmap is computed only on numerical features.
# Categorical variables were excluded (or encoded) to ensure valid correlation computation.
# The heatmap helps identify relationships between features and the target variable.

```

# (c) Age vs Heart Disease

```python

 
sns.boxplot(x='heart_disease', y='age', data=df)
plt.title("Age vs Heart Disease")
plt.show()
```

# TASK-3. Data Preprocessing

```python


from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Handle missing values
df.fillna(df.median(numeric_only=True), inplace=True)

# Separate features and target
X = df.drop('heart_disease', axis=1)
y = df['heart_disease']

# One-hot encoding
X = pd.get_dummies(X, drop_first=True)

# Scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, stratify=y, random_state=42
)

# Median used → robust to outliers
# StandardScaler → improves model performance
```

# TASK4. Model Training

```python

 
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier

# Models
dt = DecisionTreeClassifier(random_state=42)
rf = RandomForestClassifier(random_state=42)
gb = GradientBoostingClassifier(random_state=42)

# Train
dt.fit(X_train, y_train)
rf.fit(X_train, y_train)
gb.fit(X_train, y_train)
```

# TASK5. Model Evaluation

```python


from sklearn.metrics import classification_report, confusion_matrix

models = {
    "Decision Tree": dt,
    "Random Forest": rf,
    "Gradient Boosting": gb
}

for name, model in models.items():
    print(f"\n{name}")
    y_pred = model.predict(X_test)
    
    print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
    print("Classification Report:\n", classification_report(y_test, y_pred))
```

# TASK 6. Hyperparameter Tuning

```python


from sklearn.model_selection import GridSearchCV

param_grid = {
    'n_estimators': [50, 100],
    'learning_rate': [0.01, 0.1],
    'max_depth': [3, 5]
}

grid = GridSearchCV(
    GradientBoostingClassifier(random_state=42),
    param_grid,
    cv=3,
    scoring='f1'
)

grid.fit(X_train, y_train)

print("Best Parameters:", grid.best_params_)
```

# Compare Performance

```python

 
best_model = grid.best_estimator_

y_pred_tuned = best_model.predict(X_test)

print("Tuned Model Performance:\n")
print(confusion_matrix(y_test, y_pred_tuned))
print(classification_report(y_test, y_pred_tuned))

```
